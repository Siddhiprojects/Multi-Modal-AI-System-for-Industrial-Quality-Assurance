```python
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
#tanh
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
#ReLU
def relu(x):
    return max(0,x)
```

```python
#Leaky ReLU
def leaky_relu(x):
    return max(0.1*x,x)
```
